```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv("../../../res/car-data-preprocessed.csv")
new_df = df.copy()

#dropping unnecessaries
new_df.drop(labels= ['Car_Name', 'Car_Name_Numerical'], axis= 1, inplace= True) 
sns.pairplot(new_df, diag_kind= 'kde', kind='reg', plot_kws={'line_kws':{'color':'blue'}})
```

```python
sns.lmplot(data=new_df, x="Year", y="Selling_Price")
new_df["Selling_Price"].corr(new_df["Year"])
```

```python
sns.lmplot(data=new_df, x="Present_Price", y="Selling_Price")
new_df["Selling_Price"].corr(new_df["Present_Price"])
```

```python
sns.lmplot(data=new_df, x="Kms_Driven", y="Selling_Price")
new_df["Selling_Price"].corr(new_df["Kms_Driven"])
```

```python
sns.lmplot(data=new_df, x="Fuel_Type", y="Selling_Price")
new_df["Selling_Price"].corr(new_df["Fuel_Type"])
```

```python
sns.lmplot(data=new_df, x="Seller_Type", y="Selling_Price")
new_df["Selling_Price"].corr(new_df["Seller_Type"])
```

```python
sns.lmplot(data=new_df, x="Transmission", y="Selling_Price")
new_df["Selling_Price"].corr(new_df["Transmission"])
```

```python
sns.lmplot(data=new_df, x="Owner", y="Selling_Price")
new_df["Selling_Price"].corr(new_df["Owner"])
```

```python
#HeatMap
new_df = pd.get_dummies(new_df, drop_first=True)
plt.figure(figsize=(14,12))
sns.heatmap(new_df.corr(), annot= True, cmap= 'YlGnBu', vmin= -1 , vmax= 1, linecolor='black', linewidths=2)
```
